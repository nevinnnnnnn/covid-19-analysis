```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline
```

```python
covid_df = pd.read_csv("covid19_Confirmed_dataset.csv")
covid_df.drop(["Lat", "Long", "Province/State"], axis=1, inplace=True)
```

```python
covid_grouped = covid_df.groupby("Country/Region").sum()
```

```python
print("Shape:", covid_grouped.shape)
covid_grouped.head()
```

```python
countries = ['India', 'US', 'Italy', 'Spain', 'China']
for country in countries:
    covid_grouped.loc[country].plot(label=country)
```

```python
plt.legend()
plt.title("COVID-19 Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
total_cases = covid_grouped.iloc[:, -1]  # latest available data
total_cases.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Countries with Most Confirmed Cases")
plt.xlabel("Country")
plt.ylabel("Total Cases")
plt.show()
```

```python
covid_daily = covid_grouped.diff(axis=1)
covid_daily.fillna(0, inplace=True)

for country in countries:
    covid_daily.loc[country].plot(label=country)
```

```python
plt.legend()
plt.title("Daily New COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
max_infection_rates = covid_daily.max(axis=1)
covid_max_df = pd.DataFrame(max_infection_rates)
covid_max_df.columns = ['Max Infection Rate']
covid_max_df.head()
```
